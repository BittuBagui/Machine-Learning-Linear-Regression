```python
def remove_outlier(df,col):
    mean = df[col].mean()
    q1 = df[col].quantine[0.25]
    q2 = df[col].quantine[0.5]
    q3 = df[col].quantine[0.75]
    #sorted(col)
    #q1, q3 = np.percentile(col, [25,75])
    IQR = q3 - q1
    LW = q1 - (1.5 * IQR)
    UW = q3 + (1.5 *IQR)
    #return LW, UW
#for column in df.drop(['\tEmail','Address', 'Avatar' ], axis = 1):
    #LW, UW = remove_outlier(df[column])
    df.loc[df[col] > UW,col]
    df.loc[df[col] < LW,col]
    return df
```

```python
# Train Test Split
```

```python
# X = All indipendent variable
# Y = Terget Variable
```

```python
X = df[['Time on App', 'Time on Website', 'Length of Membership']]
y = df['Yearly Amount Spent']
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.3, random_state = 12)
```

```python
X_train.tail()
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train) #we always use x_train, y_train to fit the model
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2_score(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```
